# SYB Network Experience

Welcome to the SYB (Sybil Resistance) Network! This interactive notebook lets you experience how everything works in sybil resistance network with a simple UI.

## What is SYB?
SYB is a decentralized protocol that uses social vouching and algorithmic scoring to prevent sybil attacks. Users deposit funds, vouch for others they trust, and the network calculates reputation scores using algorithms like PageRank.

## How to Use This Notebook
1. **Run the network setup cells below**: This would create a random network of users and relationships.
2. **Perform Actions**: Deposit funds, vouch for others, withdraw, etc.
3. **Observe Changes**: Watch how your actions affect scores and network topology
4. **Process Batches**: Periodically forge batches to update scores
