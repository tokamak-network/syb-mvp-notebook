# SYB Network Experience

Welcome to the SYB (Sybil Resistance) Network! This interactive notebook lets you experience how everything works in a sybil resistance network with a dynamic, real-world simulation.

## What is SYB?
SYB is a decentralized protocol that uses social vouching and algorithmic scoring to prevent sybil attacks. Users deposit funds, vouch for others they trust, and the network calculates reputation scores using algorithms like PageRank.

## Key Features
- **Automatic Batch Forging**: Batches are automatically processed when the queue reaches batch size
- **Real-Time Simulation**: Every 3 seconds, a random user transaction is simulated
- **Live UI Updates**: Transaction queue, batch processing, and network graph update automatically
- **Visual Monitoring**: Watch transactions flow from queue â†’ batch processing â†’ network update

## How to Use This Notebook

3. **Watch the Simulation**: Observe automatic transactions being generated and processed
4. **Monitor the Queue**: See pending transactions in real-time
